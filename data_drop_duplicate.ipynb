## Bedbugs
Fichier qui : 
   - élimine les données en double.
   - crée une feature request_nb qui indique si la requete à été faite en 1 en 2 ou en 3... par l'utilisateur.
   - créé un fichier csv : pricing_requests_clean.csv

```python
import pandas as pd
```

```python
PATH = '.'
all_requests = pd.read_csv(PATH + '//all_our_requests_done.csv')
all_requests.loc.__setitem__((0, ('request_nb')), 1)
c = 1
for i in range(1, len(all_requests)):
    if all_requests.avatar_id.loc[i] == all_requests.avatar_id.loc[i-1] :
        c= c+1
        all_requests.loc.__setitem__((i, ('request_nb')), int(c))
        
    else :
        c = 1 
        all_requests.loc.__setitem__((i, ('request_nb')), int(c))

pricing_requests = pd.read_csv(PATH + '//pricing_requests_done.csv')
```

```python
pricing_requests = pricing_requests.merge(all_requests , on=['avatar_id', 'date', 'city', 'language', 'mobile'])
pricing_requests.drop('id_request', axis=1, inplace = True)
pricing_requests
```

```python
pricing_requests = pricing_requests.drop_duplicates(subset=['hotel_id', 'price', 'stock', 'city', 'date','language', 'mobile', 'request_nb'], keep='first', inplace=False)
```

```python
pricing_requests
```

```python
pricing_requests.to_csv('pricing_requests_clean.csv', index=False)
```
